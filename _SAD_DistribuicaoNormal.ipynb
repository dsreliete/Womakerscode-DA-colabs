# Cálculo de probabilidades

1. Qual a probabilidade de um aluno tirar nota abaixo de 12?
2. Qual a probabilidade de um aluno tirar nota entre 8 e 12?
3. Qual a probabilidade de um aluno tirar nota acima de 18?

```python
media = df.G3.mean()
desvio = df.G3.std()
print(media,desvio)
```

Tabela Z

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

tabela_normal_padronizada = pd.DataFrame(
    [],
    index=["{0:0.2f}".format(i / 100) for i in range(0, 400, 10)],
    columns = ["{0:0.2f}".format(i / 100) for i in range(0, 10)])

for index in tabela_normal_padronizada.index:
    for column in tabela_normal_padronizada.columns:
        Z = np.round(float(index) + float(column), 2)
        tabela_normal_padronizada.loc[index, column] = "{0:0.4f}".format(norm.cdf(Z))

tabela_normal_padronizada.rename_axis('Z', axis = 'columns', inplace = True)

tabela_normal_padronizada
```

```python
#pergunta 1, menor que 12
from scipy.stats import norm
#calcula a área até o valor de z
z = (12 - media)/desvio
print(z,norm.cdf(z))
```

```python
#pergunta 2, entre 8 e 12
z8 = (8 - media)/desvio
z12 = (12 - media)/desvio
p8 = norm.cdf(z8)
p12 = norm.cdf(z12)
print(p8,p12,p12-p8)
```

```python
#pergunta 3, maior que 18
z = (18 - media)/desvio
prob = norm.cdf(z)
print(prob,1 - prob)
```
