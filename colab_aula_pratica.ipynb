```python
import pandas as pd
import numpy as np
```

```python
# Simulando o banco de dados com notas de 50 alunas
np.random.seed(42)
dados = {
    'Nome': [f'Aluna_{i}' for i in range(1, 51)],
    'Prova 1': np.random.normal(6, 2, 50).round(1),
    'Prova 2': np.random.normal(7, 1.5, 50).round(1),
    'Projeto Final': np.random.normal(8, 1, 50).round(1)
}

# Criando um DataFrame
df = pd.DataFrame(dados)

# Garantindo que as notas estão no intervalo de 0 a 10
df['Prova 1'] = df['Prova 1'].clip(0, 10)
df['Prova 2'] = df['Prova 2'].clip(0, 10)
df['Projeto Final'] = df['Projeto Final'].clip(0, 10)

# Visualizando os dados
print(df.head())
```

Diferentes Tipos de Amostragem

* Amostragem Aleatória Simples: Todos os indivíduos da população têm a mesma chance de serem selecionados. Isso garante que a amostra seja representativa da população.

* Amostragem Sistemática: Selecionamos os indivíduos em intervalos regulares a partir de uma lista ordenada. Por exemplo, escolheríamos cada 5ª aluna da lista de todas as alunas.

* Amostragem Estratificada: A população é dividida em grupos (estratos) com características semelhantes, como o nível de nota, e uma amostra aleatória é tirada de cada grupo. Isso garante que todos os grupos sejam representados na amostra.

* Amostragem por Conveniência: Selecionamos os indivíduos que estão mais facilmente disponíveis, como escolher apenas as alunas que estão presentes no dia da aula. Esse método é menos confiável, pois pode introduzir vieses.

```python
# Amostra Aleatória Simples (10 alunas)
amostra_aleatoria_simples = df.sample(10)
print("Amostra Aleatória Simples:\n", amostra_aleatoria_simples)

# Amostra Sistemática (cada 5ª aluna)
amostra_sistematica = df.iloc[::5]
print("\nAmostra Sistemática (cada 5ª aluna):\n", amostra_sistematica)

# 1) Dividindo a população em estratos (notas acima e abaixo da média) para Amostragem Estratificada
media_prova1 = df['Prova 1'].mean()
grupo_abaixo_media = df[df['Prova 1'] < media_prova1]
grupo_acima_media = df[df['Prova 1'] >= media_prova1]

# 2) Selecionando 5 alunas de cada grupo
amostra_estratificada = pd.concat([grupo_abaixo_media.sample(5), grupo_acima_media.sample(5)])
print("\nAmostra Estratificada:\n", amostra_estratificada)

# Amostragem por Conveniência (por exemplo, as 5 primeiras alunas da lista)
amostra_conveniencia = df.head(5)
print("\nAmostra por Conveniência:\n", amostra_conveniencia)
```

```python
# Média
media = df[['Prova 1', 'Prova 2', 'Projeto Final']].mean()
print(f"Médias:\n{media}\n")
```

```python
# Mediana
mediana = df[['Prova 1', 'Prova 2', 'Projeto Final']].median()
print(f"Medianas:\n{mediana}\n")
```

```python
# Moda
moda = df[['Prova 1', 'Prova 2', 'Projeto Final']].mode().iloc[0]
print(f"Moda:\n{moda}\n")
```

* Amplitude
A amplitude é a diferença entre a maior e a menor nota. Ela nos dá uma ideia do intervalo total das notas.

* Variância
A variância mede o quão distante as notas estão da média. Uma variância alta indica que as notas estão bem espalhadas, e uma variância baixa indica que as notas estão mais concentradas perto da média.

* Desvio Padrão
O desvio padrão é a raiz quadrada da variância e também mede a dispersão, mas em uma unidade que é a mesma das notas. Assim, é mais fácil de interpretar em relação à média.


*  Amplitude: 7.6 — A diferença entre a maior e a menor nota.
*  Variância: 3.45 — Mede o quão dispersas as notas estão em relação à média.
*  Desvio Padrão: 1.86 — Indica, em média, o quanto as notas se afastam da média.

```python
# Calculando amplitude, variância e desvio padrão para as notas da Prova 1

# Amplitude
amplitude = df['Prova 1'].max() - df['Prova 1'].min()

# Variância
variancia = df['Prova 1'].var()

# Desvio Padrão
desvio_padrao = df['Prova 1'].std()

amplitude, variancia, desvio_padrao
```

Histograma e Distribuição Normal

* Um histograma é uma representação gráfica que mostra como os dados estão distribuídos em diferentes intervalos (ou "bins").
* Quando os dados seguem uma distribuição normal (curva de sino), o histograma terá uma forma simétrica, com mais dados concentrados em torno da média e caindo gradualmente nas extremidades.
* Se você observar um histograma com uma forma próxima de uma curva de sino, pode inferir que os dados têm uma distribuição normal.

```python
# Histograma das notas
df['Prova 1'].plot(kind='hist', alpha=0.7, bins=10, color='blue')
plt.title("Distribuição das Notas da Prova 1")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

Gráfico de Dispersão e Distribuições

* Um gráfico de dispersão mostra a relação entre duas variáveis. Ele é usado para observar correlações e tendências.
* Embora não seja usado diretamente para visualizar uma distribuição normal ou binomial, o gráfico de dispersão pode indicar a existência de uma relação linear ou não-linear entre variáveis.

```python
# Gráfico de dispersão para ver a correlação entre Prova 2 e Projeto Final
plt.scatter(df['Prova 2'], df['Projeto Final'])
plt.title("Correlação entre Prova 2 e Projeto Final")
plt.xlabel("Prova 2")
plt.ylabel("Projeto Final")
plt.show()
```
