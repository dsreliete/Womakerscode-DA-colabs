# Cálculo de probabilidades

1. Qual a probabilidade de um vinho ter o teor de alcool acima de 9.5?
2. Qual a probabilidade de um vinho ter o teor de alcool entre 10 e 12?
3. Qual a probabilidade de um vinho ter o teor de alcool menor que 8.5?

```python
from scipy.stats import norm
media = df.alcohol.mean()
desvio = df.alcohol.std()
```

```python
#pergunta 1
z = (9.5 - media)/desvio
prob = (1 - norm.cdf(z)) * 100
print(f'A probabilidade de um vinho ter o teor de alcool acima de 9.5 é de {prob:,.2f}%')
```

```python
#pergunta 2
z1 = (10 - media)/desvio
z2 = (12 - media)/desvio
prob1 = norm.cdf(z1)
prob2 = norm.cdf(z2)
prob = (prob2 - prob1)*100
print(f'A probabilidade de um vinho ter o teor de alcool entre 10 e 12 é de {prob:,.2f}%')
```

```python
#pergunta 3
z = (8.5 - media)/desvio
prob = norm.cdf(z) * 100
print(f'A probabilidade de um vinho ter o teor de alcool menor que 8.5 é de {prob:,.2f}%')
```
